```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('Bike Sales.csv')

# Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())


# Explore the structure of the dataset
print("\nDataset Info:")
print(df.info())

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# Clean the dataset by dropping missing values
df_cleaned = df.dropna()

#  basic statistics of numerical columns
print("\nBasic Statistics:")
print(df_cleaned.describe())

# Perform grouping: Average revenue per product category
grouped_data = df_cleaned.groupby('Product_Category')['Revenue'].mean()
print("\nAverage Revenue per Product Category:")
print(grouped_data)


# Line Plot - Revenue Over Time
plt.figure(figsize=(10,5))
sns.lineplot(x=df_cleaned['Year'], y=df_cleaned['Revenue'])
plt.title('Revenue Over Time')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.show()

# Bar Chart - Product Category Distribution
plt.figure(figsize=(10,5))
df_cleaned['Product_Category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Histogram - Revenue Distribution
plt.figure(figsize=(10,5))
df_cleaned['Revenue'].hist(bins=30, color='orange', edgecolor='black')
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()

# Scatter Plot - Revenue vs. Profit
plt.figure(figsize=(10,5))
sns.scatterplot(x=df_cleaned['Revenue'], y=df_cleaned['Profit'], hue=df_cleaned['Product_Category'])
plt.title('Revenue vs. Profit')
plt.xlabel('Revenue')
plt.ylabel('Profit')
plt.show()
```
